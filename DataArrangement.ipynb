```python
import os
from datetime import datetime
import shutil
from zipfile import ZipFile
```

```python

# Variables
zipPath = '../Data/Zip Data'
Datapath = '../Data/Saint Martin extracted'
Savepath = '../Data/Saint Martin'

fileName = []
harmonizedFile = []
fileCount = 0
harmonizedFileCount = 0

# A dictionary to store the month name 
monthList = {
    1: '1. Jan',
    2: '2. Feb',
    3: '3. Mar',
    4: '4. Apr',
    5: '5. May',
    6: '6. Jun',
    7: '7. July',
    8: '8. Aug',
    9: '9. Sep',
    10: '10. Oct',
    11: '11. Nov',
    12: '12. Dec'
}

allFile = {}
```

```python
# Find all files in the Data folder
# If the file ends with .tif, add it to the fileName list
# If the file ends with _harmonized.tif, add it to the harmonizedFile list
# Print the file name and the number of files found

for root, dirs, files in os.walk(Datapath):
    for file in files:
        allFile[file] = os.path.join(root, file)
        if file.endswith('.tif'):
            if file.endswith('_harmonized.tif'):
                harmonizedFile.append(file)
                harmonizedFileCount += 1
                print("Harmonized File " + str(harmonizedFileCount) + ": " + harmonizedFile[-1])
            else:
                fileName.append(file.removesuffix('.tif'))
                fileCount += 1
                print("File " + str(fileCount) + ": " + fileName[-1])    
    print('-'*50)
print('Found ' + str(fileCount) + ' files')
print('Found ' + str(harmonizedFileCount) + ' harmonized files')
print('Found ' + str(len(allFile)) + ' files in total')
```

```python
print(allFile)
```

```python
# Copy the harmonized files to the Savepath
# The file name is the date of the image
# The file is copied to the folder of the year and month of the image
# If the folder does not exist, create it
# Print the file name and the path it was copied to
# Harmonized files only
count = 0
for i in allFile.keys():
    if i.startswith('2'):
        date = i.split('_')[0]
        date =datetime.strptime(date, "%Y%m%d")
        year = date.year
        month = date.month
        month = monthList[month]
        path = f'{Savepath}/{year}/{month}'
        if not os.path.exists(path):
            os.makedirs(path)
        shutil.copy(allFile[i],path)
        print(f'{i} Copied to {path}')
        count += 1
print(f'-{count}-'*50)        
print(f'{count} files copied')

```
