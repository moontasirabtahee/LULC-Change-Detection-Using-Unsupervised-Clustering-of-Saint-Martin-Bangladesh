```python
import os
from datetime import datetime
import shutil
from zipfile import ZipFile
```

```python
Datapath = 'E:\LULC Change Detection Using Unsupervised Clustering of Saint Martin Bangladesh\Data File\Saint Martin Extracted'
Savepath = 'E:\LULC Change Detection Using Unsupervised Clustering of Saint Martin Bangladesh\Data File\Saint Martin'

allFile = {}

# A list to store the file name
udmFile = []
harmonizedFile = []
metadata = []
udmFileCount = 0
harmonizedFileCount = 0
metadataCount = 0   


# A dictionary to store the month name 
monthList = {
    1: '1. Jan',
    2: '2. Feb',
    3: '3. Mar',
    4: '4. Apr',
    5: '5. May',
    6: '6. Jun',
    7: '7. July',
    8: '8. Aug',
    9: '9. Sep',
    10: '10. Oct',
    11: '11. Nov',
    12: '12. Dec'
}

```

```python
# Find all files in the Data folder
# If the file ends with .tif, add it to the fileName list
# If the file ends with _harmonized.tif, add it to the harmonizedFile list

# If the file ends with metadata.xml, add it to the metadata list

# Print the file name and the number of files found

for root, dirs, files in os.walk(Datapath):
    for file in files:
        if file in Savepath:
            continue
        if file.endswith('.tif'):
            if file.endswith('_harmonized.tif'):
                harmonizedFile.append(file)
                harmonizedFileCount += 1
                allFile[file] = os.path.join(root, file)
                print("Harmonized File " + str(harmonizedFileCount) + ": " + harmonizedFile[-1])
            
            else:
                udmFile.append(file.removesuffix('.tif'))
                udmFileCount += 1
                allFile[file] = os.path.join(root, file) 
                print("File " + str(udmFileCount) + ": " + udmFile[-1]) 
        elif file.endswith('metadata.xml'):
            metadata.append(file)
            metadataCount += 1
            allFile[file] = os.path.join(root, file)
            
            print("Metadata " + str(metadataCount) + ": " + metadata[-1])
            print(allFile[file])
            
    print('-'*50)
print('Found ' + str(udmFileCount) + ' files')
print('Found ' + str(harmonizedFileCount) + ' harmonized files')
print('Found ' + str(metadataCount) + ' metadata files')
print('Found ' + str(len(allFile)) + ' files in total')
```

```python
# print the all files in line by line
# print the number of files found
# print the number of harmonized files found
# print the number of metadata files found
# print the number of files found in total

for file in allFile:
    print(file)
print('Found ' + str(udmFileCount) + ' files')
print('Found ' + str(harmonizedFileCount) + ' harmonized files')
print('Found ' + str(metadataCount) + ' metadata files')
print('Found ' + str(len(allFile)) + ' files in total')
```

```python
# Copy the harmonized files to the Savepath
# The file name is the date of the image
# The file is copied to the folder of the year and month of the image
# If the folder does not exist, create it
# Print the file name and the path it was copied to
# Harmonized files only
count = 0
for i in allFile.keys():
    if i.startswith('2'):
        date = i.split('_')[0]
        date =datetime.strptime(date, "%Y%m%d")
        year = date.year
        month = date.month
        month = monthList[month]
        path = f'{Savepath}/{year}/{month}'
        if not os.path.exists(path):
            os.makedirs(path)
        # if the file is already in the path, skip it
        if os.path.exists(f'{path}/{i}'):
            print(f'{i} already exists in {path}\n')
            continue
        shutil.copy(allFile[i],path)
        print(f'{count} :  {i} Copied to {path}\n')
        count += 1
print(f'-{count}-'*50)        
print(f'{count} files copied')

```
